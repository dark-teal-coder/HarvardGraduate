### Unit tests for justify()

<div class="alert alert-block alert-danger">
Now run the Unit Tests to check your work
</div>

````python
def validate_justify(before: str):
    after = right_justify(before)
    assert len(after) == 70, f"Wrong length {len(after)} after justifying {before}."
    if (before): 
        assert before == after[-len(before):], f"Should find '{before}' on right end."
    length = 70 - len(before)
    if length > 0:
        assert ' '*length == after[:length], f"Should find blanks on left end."


def validate_suite():
    validate_justify(' ')
    validate_justify('one')
    validate_justify('two')
    validate_justify('three')
    validate_justify('Monty')
    validate_justify('Project Oxygen shocked everyone by concluding that, among the eight...')
    validate_justify('')

    print('Success!')

validate_suite()
````

## Solution

````python
# right_justify('monty') should return a string of length 70, ending with monty:
# returns: '                                                                 monty'
def right_justify(text):
    return ' '*(70 - len(text)) + text

validate_suite()
````

## Replace constants (4) with variables

There are three instances of 4, which stand for three different things

We say that "4 is a magic number" - in fact, it is three different magic numbers

The size of a box
```python
def beam() -> str:
    return '-'*4 + '+'

def boxes() -> str:
    return ...(spacers()*4 + row())...
```

The number of columns
```python
# row() returns '+----+----+----+----+\n'
def row() -> str:
    return '+' + beam()*4 + '\n'

# spaces() returns '|    |    |    |    |\n'
def spacers() -> str:
    return '|' + (' '*... + '|')*4 + '\n'
```

The number of rows
```python
# boxes
def boxes() -> str:
    return row() + (spacers()*... + row())*4
```

## Next step: make some of these constants variable

````python
# beam(2) returns beam of size 2 '--+'
def beam(size: int) -> str:
    return '-'*size + '+'

# row(2, 5) returns '+--+--+--+--+--+\n'
def row(size: int, count: int) -> str:
    return '+' + beam(size)*count + '\n'

# spaces(2, 3) returns '|  |  |  |\n'
def spacers(size: int, count: int) -> str:
    return '|' + (' '*size + '|')*count + '\n'

# boxes
def boxes(size: int, count: int) -> str:
    return row(size, count) + (spacers(size, count)*size + row(size, count))*count

size  = 4       # Size of each 'square'
count = 3       # Number of rows and columns

print(boxes(4, 3))
print(boxes(5, 2))
print(boxes(2, 5))
````

## Third step: make size, rows, columns all variable, and document

I have combined some routines and added documentation

````python
# Grid.py
#
# Jeff Parker, July 2017
#
# Draw a grid of shapes

# Build an edge: row or spacers
#
# All rows are the same basic pattern
# +---+---+---+---+
#
# |   |   |   |   |
# 
def build_edge(marker: str, filler: str, edge_len: int, num_cols: int) -> str:
    "Build an edge"
    # unit =  ---+  
    unit = filler * edge_len + marker

    # result = + ---+ ---+ ---+ ---+ 
    return marker + num_cols * unit + "\n"


# top_edge - return the top edge
# Return
# +---+---+---+---+
# 
def top_edge(edge_len: int, num_cols: int) -> str:
    "Return the top edge"
    return build_edge("+", "-", edge_len, num_cols)


# row - return a row
# Return
# |   |   |   |   |
# 
def row(edge_len: int, num_cols: int) -> str:
    "return an internal row"
    return build_edge("|", " ", edge_len, num_cols)


# Block - Combine rows with one edge
# Return
# |   |   |   |   |
# |   |   |   |   |
# |   |   |   |   |
# +---+---+---+---+
#
def block(edge_len: int, num_cols: int) -> str:
    # unit = |   |   |   |   |
    unit = row(edge_len, num_cols)

    # tail = +---+---+---+---+
    tail = top_edge(edge_len, num_cols)

    result = edge_len*unit + tail
    return result


# grid 
# Use the components above to return a grid
#
def grid(edge_len: int, num_cols: int, num_rows: int) -> str:
    head = top_edge(edge_len, num_cols)
    unit = block(edge_len, num_cols)
    result = head + num_rows * unit
    return result

# Construct a grid and print it
print(grid(4, 2, 3))
````
